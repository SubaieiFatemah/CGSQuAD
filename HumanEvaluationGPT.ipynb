

### GPT-4 - Filling

```python
import pandas as pd
gpt4_eval_df = pd.read_csv('similarity_evaluation_GPT_4.csv')#semantic similarity evaluation file for gpt4
```

```python
for index, row in gpt4_eval_df.iterrows():
  generated_response = row['generated_response']
  match = gpt3_eval_df[gpt3_eval_df['generated_response'].str.startswith(generated_response[:35])]

  if not match.empty:
    gpt4_eval_df.at[index, 'human evaluation score'] = match.iloc[0]['human evaluation score']
```

```python
gpt4_eval_df.loc[gpt4_eval_df['generated_response'].str.contains('النص لا يوفر معلومات'), 'human evaluation score'] = 0
gpt4_eval_df.loc[gpt4_eval_df['generated_response'].str.contains('عذرًا'), 'human evaluation score'] = 0
gpt4_eval_df.loc[gpt4_eval_df['generated_response'].str.contains('لا أعلم'), 'human evaluation score'] = 0
gpt4_eval_df.loc[gpt4_eval_df['generated_response'].str.contains('لا يوجد'), 'human evaluation score'] = 0
gpt4_eval_df.loc[gpt4_eval_df['generated_response'].str.contains('المعلومات المتاحة لا'), 'human evaluation score'] = 0

def check_for_match(generated_response, corresponding_response):
  return generated_response.find(corresponding_response) != -1
# Update the code to return 1 if found and leave it as is if not found
gpt4_eval_df['human evaluation score'] = gpt4_eval_df.apply(lambda row: 1 if check_for_match(row['generated_response'], row['corresponding_response']) else row['human evaluation score'], axis=1)


def starts_with(string):
  return string.lower().startswith('نعم')

# Set the human evaluation score to 1 if the generated response and corresponding response both start with 'yes'
gpt4_eval_df['human evaluation score'] = gpt4_eval_df.apply(lambda row: 1 if starts_with(row['generated_response']) and starts_with(row['corresponding_response']) else row['human evaluation score'], axis=1)


def starts_with(string):
  return string.lower().startswith('لا ')

# Set the human evaluation score to 1 if the generated response and corresponding response both start with 'no'
gpt4_eval_df['human evaluation score'] = gpt4_eval_df.apply(lambda row: 1 if starts_with(row['generated_response']) and starts_with(row['corresponding_response']) else row['human evaluation score'], axis=1)

gpt4_eval_df['generated_response'] = gpt4_eval_df['generated_response'].str.replace('،', '')
gpt4_eval_df['generated_response'] = gpt4_eval_df['generated_response'].str.replace(':', '')

gpt3_eval_df['generated_response'] = gpt3_eval_df['generated_response'].str.replace('،', '')
gpt3_eval_df['generated_response'] = gpt3_eval_df['generated_response'].str.replace(':', '')

gpt4_eval_df.loc[gpt4_eval_df['similarity_score'] >= 0.53, "human evaluation score"] = 1
gpt4_eval_df.loc[gpt4_eval_df['similarity_score'] < 0, "human evaluation score"] = 0 #summarizes all "I do not know" cases and wrong answers

gpt4_eval_df
```

```python
#count empty cells to evaluate them manually
empty_value_count = gpt4_eval_df.isna().sum().sum()
print("Empty cells that require manual human evaluation: ", empty_value_count)

done_eval = ((len(gpt4_eval_df)-empty_value_count)/len(gpt4_eval_df))*100
print("Human evaluation percentage for GPT-4 is done by: ", done_eval, '%')

missing_eval = (empty_value_count/len(gpt4_eval_df))*100
print("Human evaluation percentage for GPT-4 is missing by: ", missing_eval, '%')
```

```python
#evaluate missing rows one by one
gpt4_eval_df.to_excel('gpt4_evaluation.xlsx', index=False)
#now download, manually evaluate, upload, then calculate final human evaluation
```
